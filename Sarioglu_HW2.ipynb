```python
bp_readings = [120, 137, 151, 123, 162, 173, 127, 154]

def normalize_value(reading):
  min_value = min(reading)
  max_value = max(reading)

  normalized = []

  for x in reading:
    normalized_value = (x-min_value)/(max_value-min_value)
    normalized.append(normalized_value)

  return normalized

normalized_bp_readings = normalize_value(bp_readings)
print(normalized_bp_readings)
```
